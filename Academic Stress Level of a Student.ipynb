```python
from sklearn.cluster import KMeans

X=data.drop(['Timestamp'],axis=1)
```

```python
X.replace({
    'Your Academic Stage':{'undergraduate':0, 'high school':1, 'post-graduate':2},
    'Study Environment':{'Noisy':0, 'Peaceful':1, 'disrupted':2},
    'What coping strategy you use as a student?':{'Analyze the situation and handle it with intellect':0,
       'Social support (friends, family)':1,
       'Emotional breakdown (crying a lot)':2},
    'Do you have any bad habits like smoking, drinking on a daily basis?':{'No':0, 'prefer not to say':1, 'Yes':2}

},inplace = True)
```

```python
print(X.to_string())
```

```python
model = KMeans(n_clusters = 2)
model.fit(X)
```

```python
"""
'Your Academic Stage':{'undergraduate':0, 'high school':1, 'post-graduate':2},
    'Study Environment':{'Noisy':0, 'Peaceful':1, 'disrupted':2},
    'What coping strategy you use as a student?':{'Analyze the situation and handle it with intellect':0,
       'Social support (friends, family)':1,
       'Emotional breakdown (crying a lot)':2},
    'Do you have any bad habits like smoking, drinking on a daily basis?':{'No':0, 'prefer not to say':1, 'Yes':2}
"""


A =int(input("Your Academic Stage -{'graduate':0, 'high school':1, 'post-graduate':2 :- "))
B=int(input("Peer pressure [ 1 To 5] :- "))
C =int(input("Academic pressure from your home [1 To 5] :- "))
D = int(input("Study Environment -{'Noisy':0, 'Peaceful':1, 'disrupted':2} :- "))
E=int(input("What coping strategy you use as a student? -{'Analyze the situation and handle it with intellect':0,'Social support (friends, family)':1'Emotional breakdown (crying a lot)':2} :- "))
F=int(input("Do you have any bad habits like smoking, drinking on a daily basis? - {'No':0, 'prefer not to say':1, 'Yes':2} :- "))
G=int(input("What would you rate the academic  competition in your student life - [1 To 5] :- "))
H=int(input("Rate your academic stress index -[1 To 5] :- "))

user_input=([A,B,C,D,E,F,G,H])

y_pred = model.predict([user_input])

if ( A == 1 and y_pred == [1]):
  print("As you are a High School Student : Don’t take studies as a burden. Focus on learning, not just on marks. Keep a balance between books and hobbies—play games, draw, dance, or listen to music. Don’t compare yourself with others; everyone has their own path. If you feel pressure, talk to your teachers or parents—sharing always reduces stress.")
elif (A == 0  and y_pred == [1] ):
  print("As you are a Graduate  Student : This is the stage where you start worrying about your career, but remember: skills are more important than grades alone. Join clubs, participate in activities, and explore what you enjoy. Manage your time smartly and don’t push yourself into burnout. If things feel heavy, seek guidance from mentors or friends—college is not just about studying, it’s about discovering yourself.")
elif (A == 2 and y_pred == [1]) :
  print("As you are a Post-Graduate Student : Post-graduation brings pressure from research, deadlines, and the future. Remember, you’re not alone—many go through the same stress. Take breaks, stay physically active, and don’t let perfectionism consume you. Network with peers, share your struggles, and celebrate small progress. It’s okay to rest—it’s not wasted time, it’s recharging for better focus.")
else:
  print("Peaceful students become powerful learners And Your one of them!!")
```
