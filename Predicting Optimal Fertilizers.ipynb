```python
df.replace({
    'Crop_Type':{'Rice':0, 'Barley':1, 'Sugarcane':2, 'Soybean':3, 'Maize':4, 'Cotton':5,'Wheat':6},
    'Soil_Type' : {'Chalky':0, 'Sandy':1, 'Silty':2, 'Peaty':3, 'Loamy':4, 'Clay':5},
    'Recommended_Fertilizer':{'10-26-26':0, 'NPK':1, 'DAP':2, 'Urea':3, 'Compost':4, 'Vermicompost':5}

},inplace =True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X =df.drop("Recommended_Fertilizer",axis = 1)
y = df['Recommended_Fertilizer']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
model = RandomForestClassifier(n_estimators = 100)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score

print("confusion_matrix : ",confusion_matrix(y_test,y_pred))
print("accuracy_score : ",accuracy_score(y_test,y_pred))
```
